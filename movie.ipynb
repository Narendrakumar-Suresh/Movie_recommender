```python
import pandas as pd
```

```python
df=pd.read_excel('netflix.xlsx')
df=pd.DataFrame(df)
df.rename(columns={"Series or Movie": "Series"}, inplace=True)
df.dropna().head()
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
df['Genre'] = df['Genre'].apply(lambda x: [genre.strip() for genre in str(x).split(',')] if pd.notnull(x) else [])
mlb = MultiLabelBinarizer()
genre_encoded = mlb.fit_transform(df['Genre'])
genre_df = pd.DataFrame(genre_encoded, columns=mlb.classes_)
df = pd.concat([df, genre_df], axis=1)
df.drop(columns=['Genre'], inplace=True)
```

*Removing unused columns*

```python
remove_list=['Tags','Languages','Country Availability', 'Runtime', 'Director', 'Writer', 'Actors', 'Metacritic Score', 'Awards Received', 'Awards Nominated For',
       'Boxoffice', 'Release Date', 'Netflix Release Date', 'Production House',
       'Netflix Link', 'IMDb Link', 'Summary', 'IMDb Votes', 'Image', 'Poster',
       'TMDb Trailer', 'Trailer Site','Rotten Tomatoes Score']

for i in remove_list:
    df.drop(columns=[i], inplace=True)
final_df=df
```

```python
most_recurring_value = final_df['View Rating'].mode()[0]
most_imdb=final_df['IMDb Score'].mean()
most_hid=final_df['Hidden Gem Score'].mean()
final_df['View Rating']=final_df['View Rating'].fillna(most_recurring_value)
final_df['IMDb Score']=final_df['IMDb Score'].fillna(most_imdb)
final_df['Hidden Gem Score']=final_df['Hidden Gem Score'].fillna(most_hid)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

final_df['View Rating'] = final_df['View Rating'].apply(lambda x: [rating.strip() for rating in str(x).split(',')] if pd.notnull(x) else [])

mlb_view_rating = MultiLabelBinarizer()
view_rating_encoded = mlb_view_rating.fit_transform(final_df['View Rating'])

view_rating_df = pd.DataFrame(view_rating_encoded, columns=mlb_view_rating.classes_)
final_df = pd.concat([final_df, view_rating_df], axis=1)

final_df.drop(columns=['View Rating'], inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
final_df['Series'] = le.fit_transform(final_df['Series'])
final_df.dropna()
```

```python
# Print all the column names to ensure they match
print(final_df.columns)
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()

scaled_features = scaler.fit_transform(final_df[['Hidden Gem Score', 'IMDb Score']])
scaled_df = pd.DataFrame(scaled_features, columns=['Hidden Gem Score', 'IMDb Score'])

non_numeric_columns = final_df.select_dtypes(exclude=['float64', 'int64']).columns
scaled_df = pd.concat([scaled_df, final_df[non_numeric_columns]], axis=1)
```

**Similarity**

```python
movie_features = final_df.drop(columns=['Title'])
movie_features = movie_features.apply(pd.to_numeric, errors='coerce')
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
movie_features_scaled = scaler.fit_transform(movie_features)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(movie_features_scaled)
```

***Saving the model***

```python
import joblib

# Save the model to a file
joblib.dump(cosine_sim, 'movie_predict.pkl')
print("Model saved successfully!")
```

```python
cosine_sim = joblib.load('movie_predict.pkl')
```

***Prediction***

```python
def recommendation(title, cosine_sim=cosine_sim):
    if title not in final_df['Title'].values:
        return f"Title '{title}' not found in the dataset."
    
    idx=final_df[final_df['Title']==title].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:6]
    movie_indices = [i[0] for i in sim_scores]
    result=final_df['Title'].iloc[movie_indices].tolist()
    return result
```

```python
final_df.to_csv('final_movie_data.csv', index=False)
```

```python
# Example: Recommend movies similar to 'Joker'
recommended_movies = recommendation('Scott Pilgrim vs. the World')
print(recommended_movies)
```
